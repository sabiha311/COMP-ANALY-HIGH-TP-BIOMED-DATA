```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from google.colab import files

# Step 1: Upload files manually
uploaded = files.upload()

# Step 2: Load the matrices
matrix1 = pd.read_csv("matrix1 (1).txt", sep="\t", index_col=0)
matrix2 = pd.read_csv("matrix2 (1).txt", sep="\t", index_col=0)

# Step 3: Compute Pearson correlation matrices
corr_matrix1 = matrix1.corr(method='pearson')
corr_matrix2 = matrix2.corr(method='pearson')

# Step 4: Compute the overall Pearson correlation matrix between matrix1 and matrix2
combined_corr_matrix = matrix1.corrwith(matrix2, axis=0).to_frame(name="Pearson_Correlation")

# Function to plot and save heatmaps
def plot_heatmap(corr_matrix, title, filename):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title(title)
    plt.savefig(filename)
    plt.show()

# Generate and save heatmaps
plot_heatmap(corr_matrix1, "Pearson Correlation Matrix - Matrix 1", "correlation_matrix1.png")
plot_heatmap(corr_matrix2, "Pearson Correlation Matrix - Matrix 2", "correlation_matrix2.png")

# Step 5: Save correlation matrices as CSV
corr_matrix1.to_csv("correlation_matrix1.csv")
corr_matrix2.to_csv("correlation_matrix2.csv")
combined_corr_matrix.to_csv("overall_correlation.csv")

# Step 6: Download all generated files
files.download("correlation_matrix1.csv")
files.download("correlation_matrix2.csv")
files.download("overall_correlation.csv")
files.download("correlation_matrix1.png")
files.download("correlation_matrix2.png")


```
